3. 数组拼接（多时序数据合并）：创建两个一维数组（分别模拟两组不同时段的时序数据，如[10,11,12]和[13,14,15]），将其拼接为一个更长的一维数组。

````python
ln = [7,8,9,np.nan,6,8,10,np.nan]
arrn = np.array(ln, dtype=float, order='C')
arrn1 = np.where(np.isnan(arrn),np.nanmean(arrn),arrn)
# np.nanmean(arrn)
arrn1
np.diff(arrn1)
np.concatenate((arr1,arrn1),axis = 0)
````
