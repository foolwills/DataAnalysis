# 时间序列分析学习

第 1 周：Python 数据分析必备库 0316

- 目标：能读取、处理、简单查看数据

- 内容：


  - NumPy 数组基础操作

  - Pandas：Series、DataFrame、读取 csv/excel

  - 基础数据清洗：缺失值、重复值

- 实战：加载公开时序数据（气温/销量/股票），做简单统计（均值、最值、切片筛选）

第 2 周：Pandas 时间序列专用功能

- 目标：熟练处理“带时间”的数据，构建标准时间序列

- 内容：


  - pd.to_datetime 函数使用、时间索引创建与操作

  - 重采样 resample（日→周、周→月等时间粒度转换）

  - 滑动窗口 rolling 操作（计算移动平均）

  - 时间偏移、日期范围生成（pd.date_range）

- 实战：将普通表格数据（含日期列）转为带时间索引的标准时间序列，完成基础时间粒度转换


第 3 周：时间序列基础概念 + 可视化

- 目标：理解时间序列核心结构，能通过可视化呈现数据规律

- 内容：


  - 时间序列四大组成部分：趋势、季节性、周期性、噪声

  - 平稳性初步理解（什么是平稳序列、非平稳序列）

  - Matplotlib/Seaborn 绘制时序图（折线图、子图、趋势图）

- 实战：加载时序数据，绘制折线图、按季节拆分的子图，初步分析数据趋势和季节性


第 4 周：时间序列预处理（工程核心）

- 目标：将“脏数据”处理为可建模的“干净数据”，掌握时序预处理规范

- 内容：


  - 时序缺失值填充（时间序列专用方法：前向填充、后向填充、插值填充）

  - 异常值检测与处理（3σ原则、箱线图检测，异常值替换/删除）

  - 时序数据标准化/归一化（MinMaxScaler、StandardScaler）

  - 训练集与测试集划分（关键：不能随机打乱，按时间顺序划分）

- 实战：获取一份真实时序脏数据（含缺失值、异常值），完成完整预处理流程，输出可建模数据


第 5 周：预测评价指标 + 基础预测模型

- 目标：掌握时序预测评价标准，能使用基础方法完成简单预测

- 内容：


  - 核心评价指标：MAE（平均绝对误差）、MSE（均方误差）、RMSE（均方根误差）、MAPE（平均绝对百分比误差）

  - 基础预测方法：均值法、移动平均法、简单指数平滑法

- 实战：用3种基础模型对预处理后的时序数据做短期预测，计算各模型评价指标，对比模型效果


第 6 周：统计模型核心 —— ARIMA

- 目标：进入专业时序建模领域，掌握经典统计模型的使用

- 内容：


  - 平稳性检验：ADF检验（Augmented Dickey-Fuller Test）

  - 差分操作（一阶差分、二阶差分），将非平稳序列转为平稳序列

  - ARIMA模型原理（p、d、q参数含义）

  - pmdarima库使用（auto_arima自动选阶，简化建模流程）

- 实战：用ARIMA模型对单变量时序数据（如气温、单产品销量）做预测，对比基础模型效果


第 7 周：SARIMA / SARIMAX 季节与外生变量

- 目标：解决带季节性的商业场景预测问题，能融入外部影响因素

- 内容：


  - SARIMA模型原理（处理季节性时序数据，增加季节参数P、D、Q）

  - SARIMAX模型（加入外生变量，如节假日、天气、促销活动等）

- 实战：以电商销量为数据，融入节假日外生变量，用SARIMAX模型做销量预测，优化预测精度


第 8 周：机器学习时序特征工程

- 目标：掌握时序数据转监督学习格式的核心方法，构建有效特征

- 内容：


  - 时间特征构造：小时、星期、月份、季度、节假日标记、是否周末

  - 滞后特征（Lag特征）：前1天、前7天、前30天的历史数据作为特征

  - 窗口统计特征：滚动均值、滚动方差、滚动最值（7天、30天窗口）

  - 时序数据转监督学习格式（将序列问题转为X-y预测问题）

- 实战：对时序数据批量构造10+个有效特征，生成可用于机器学习的训练数据


第 9 周：机器学习预测模型（XGBoost/LightGBM）

- 目标：掌握工业界主流的时序预测方案，提升预测精度

- 内容：


  - 树模型基础（决策树、随机森林），理解其时序预测逻辑

  - XGBoost/LightGBM 模型原理与使用（调参基础、时序适配）

- 实战：基于第8周构造的特征，搭建XGBoost/LightGBM预测模型，完成完整的机器学习时序预测流程（特征筛选→训练→预测→评价）
