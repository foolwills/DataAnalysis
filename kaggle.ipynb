```python
import duckdb

# 连接数据库
con = duckdb.connect("ecommerce.duckdb")

tables = con.execute("SHOW TABLES").fetchall()

print(tables)
```

```python
# 新增 创建表并导入CSV
con.execute("""
CREATE TABLE olist_customers AS
SELECT * FROM read_csv_auto('data/olist/olist_customers_dataset.csv')
""")
```

```python
# 新增 创建表并导入CSV
con.execute("CREATE TABLE olist_order_payments AS SELECT * FROM read_csv_auto('data/olist/olist_order_payments_dataset.csv')")
con.execute("CREATE TABLE olist_order_reviews AS SELECT * FROM read_csv_auto('data/olist/olist_order_reviews_dataset.csv')")
con.execute("CREATE TABLE olist_orders AS SELECT * FROM read_csv_auto('data/olist/olist_orders_dataset.csv')")
con.execute("CREATE TABLE olist_products AS SELECT * FROM read_csv_auto('data/olist/olist_products_dataset.csv')")
con.execute("CREATE TABLE olist_sellers AS SELECT * FROM read_csv_auto('data/olist/olist_sellers_dataset.csv')")
con.execute("CREATE TABLE product_category_name_translation AS SELECT * FROM read_csv_auto('data/olist/product_category_name_translation.csv')")
```

```python
# 删除
con.execute("""
DROP TABLE olist
""")
```

```python
# 查询
print(con.execute("SELECT * FROM olist_customers LIMIT 5").fetchall())
```

```python
# 关闭数据库
con.close()
```
