```python
from pyecharts.charts import Map
from pyecharts.options import VisualMapOpts
```

```python
# 虚拟数据
data = [
    ("广东", 500),
    ("北京", 300),
    ("上海", 400),
    ("浙江", 350),
    ("江苏", 450),
    ("山东", 380),
    # 可继续添加其他省份数据
]
```

```python
map_chart = Map()
map_chart.add("销售量", data, "china")
# map_chart.set_global_opts(title_opts=opts.TitleOpts(title="省份销量图"))
map_chart.set_global_opts(
    visualmap_opts=VisualMapOpts(
        is_show=True,
        min_=0,
        max_=600,
        range_text=["最大值", "最小值"],
        range_color=["lightskyblue", "yellow", "orangered"],
    )
)

# 生成HTML文件
map_chart.render("province_sales.html")
```
